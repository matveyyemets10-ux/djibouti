```python
def check_input(input_str):
    numbers = []
    for item in input_str.split():
        if not item.isdigit():
            return f"Error. There is a non-value in the list: '{item}'"
        numbers.append(int(item))
    return numbers

def bubble_sort_count(numbers):
    swaps = 0
    n = len(numbers)
    for i in range(n):
        for j in range(n - i - 1):
            if numbers[j] > numbers[j + 1]:
                numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]
                swaps += 1
    return swaps

def task_4():
    input_str = input()
    result = check_input(input_str)
    if type(result) == str:
        print(result)
    else:
        swaps = bubble_sort_count(result)
        print(f"There was {swaps} swaps")
```

```python
#"5 6 23 8 12 5"
# Should print: There was 6 swaps
task_4()
```

```python
#"7 3 1 7 4 2 8 2"
# Should print: There was 14 swaps
task_4()
```
